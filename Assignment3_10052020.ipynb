```python
def play():
    subjects=(["Americans ","Indians"])
    verbs=(["play","watch"])
    objects=(["Baseball","Cricket"])

    n= [subjects[i]+" "+j+" "+k for k in objects for j in verbs for i in range(len(subjects))]
    for i in n:
        yield i
print(play())
```

```python
import numpy as np

def matrix(m,n):
    lst=[1,2,3,4,5]
    num=len(lst)
    dec,inc=[],[]
    
    if m==1:        
        l1=np.array([[[row **i for i in range(n)] for row in range(1,num+1)]])
        for i in l1:
            yield i     
    else:
        l2=np.array([[[row **(n-1-i) for i in range(n)]  for row in range(1,num+1)]])
        for i in l2:
            yield i 
            
n=int(input("Enter the no: of Column to be printed: "))             
result=int(input("Enter the choice of decreasing/Increasing :"))
print(matrix(result,n))
```
